```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
targets = data_preprocessed['Absenteeism Time in Hours'].apply(lambda x: 1 if x > data_preprocessed['Absenteeism Time in Hours'].median() else 0)
```

```python
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
```

```python
targets.value_counts()
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours', 'Daily Work Load Average',
 'Day of the Week',
 'Distance to Work'], axis=1)
```

```python
data_with_targets is data_preprocessed
```

```python
unscaled_inputs = data_with_targets.drop(['Date'], axis=1).iloc[:,:-1]
```

```python
unscaled_inputs.head()
```

```python
unscaled_inputs_without_dummies = unscaled_inputs.drop(columns=['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education'], axis=1)
dummies = unscaled_inputs[['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education']]
unscaled_inputs_without_dummies
```

```python
from sklearn.preprocessing import StandardScaler

absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs_without_dummies)
```

```python
scaled_inputs_without_dummies = absenteeism_scaler.transform(unscaled_inputs_without_dummies)
```

```python
scaled_inputs_without_dummies
```

```python
scaled_inputs = pd.DataFrame(scaled_inputs_without_dummies, columns = unscaled_inputs_without_dummies.columns)
```

```python
scaled_inputs
```

```python
scaled_inputs = pd.concat([scaled_inputs, dummies], axis=1)
```

```python
scaled_inputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size=0.8, random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

```python
model_outputs = reg.predict(x_train)
```

```python
(model_outputs == y_train).sum()/x_train.shape[0]
```
